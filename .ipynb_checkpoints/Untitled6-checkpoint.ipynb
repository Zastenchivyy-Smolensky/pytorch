```python
from sklearn.datasets import load_diabetes
from sklearn.metrics import mean_squared_error
import numpy as np
diabetes = load_diabetes()
diabetes
```

```python
import pandas as pd
pd.DataFrame(diabetes.data, columns=("age", "sex", "bmi", "map", "s1", "s2", "s3", "s4", "s5", "s6"))
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(diabetes.data, diabetes.target, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(train_X, train_y)
print(model.score(test_X, test_y))
print(model.predict(test_X))
```

```python
print(model.score(test_X, test_y))
```

```python
print(model.score(train_X, train_y))
```

```python
pred_y=model.predict(test_X)
```

```python
(pred_y-test_y)**2
```

```python
mse = mean_squared_error(test_y, pred_y)
print(np.sqrt(mse))#平均二乗誤差
```

```python
mse = np.mean((test_y-pred_y)**2)
print(mse)
```

```python
mae = np.mean(np.abs(test_y-pred_y))
print(mae)
```

```python
pred_y=model.predict(train_X)
mse = np.mean((train_y-pred_y)**2)
print(mse)#平均二乗誤差
mae = np.mean(np.abs(train_y-pred_y))
print(mae)
```
